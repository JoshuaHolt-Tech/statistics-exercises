1. A bank found that the average number of cars waiting during the noon hour at a drive-up window follows a Poisson distribution with a mean of 2 cars. Make a chart of this distribution and answer these questions concerning the probability of cars waiting at the drive-up window.

```python
x1 = range(8)
y1 = stats.poisson(2).pmf(x1)

```

```python
plt.bar(x1, y1, width=1, color='orange', edgecolor='black')
```

    - What is the probability that no cars drive up in the noon hour?
        - Answer: 13.5%

```python
print(f'The probability of {x1[0]} cars is {y1[0]:.3f}')
```

    - What is the probability that 3 or more cars come through the drive through?
        - Answer: 5.3%

```python
y1_2 = stats.poisson(2).sf(4)
print(f'The probability of greater than 3 cars is {y1_2:.3f} %')
```

    - How likely is it that the drive through gets at least 1 car?
        - Answer: 86.5%

```python
y1_3 = stats.poisson(2).sf(0)
print(f'The probability at least one car is {y1_3:.3f} %')
```

2. Grades of State University graduates are **normally distributed** with a **mean of 3.0** and a **standard deviation of .3.** Calculate the following:

```python
x2 = np.arange(1.5, 4.5,0.01)
y2 = stats.norm(loc=3,scale=.3).pdf(x2)
plt.plot(x2, y2, color='black')
```

    - What grade point average is required to be in the top 5% of the graduating class?
        - Answer: GPA = 3.4934

```python
print(stats.norm(loc=3,scale=.3).ppf(.95))
```

    - What GPA constitutes the bottom 15% of the class?
        - Answer: Less than GPA = 2.6891

```python
print(stats.norm(loc=3,scale=.3).ppf(.15))
```

    - An eccentric alumnus left scholarship money for students in the third decile from the bottom of their class. Determine the range of the third decile.
       - Answer:  2.75 - 2.83 (Mine has 20-29%)

```python
print(f'The range of the third decile is: {round(stats.norm(loc=3,scale=.3).ppf(.20),2)} to {round(stats.norm(loc=3,scale=.3).ppf(.29),2)}')
```

    - Would a student with a 2.8 grade point average qualify for this scholarship?
        - Answer: No

```python
print( f'No, {stats.norm(loc=3,scale=.3).cdf(2.8):.3f} is below third decile.')
```

    - If I have a GPA of 3.5, what percentile am I in?
        - Answer: 90 percentile

```python
print( f'You are at {stats.norm(loc=3,scale=.3).cdf(3.5)*100:.0f} % in the class')
```

6. You want to get lunch at La Panaderia, but notice that the line is usually very long at lunchtime. After several weeks of careful observation, you notice that the average number of people in line when your lunch break starts is normally distributed with a mean of 15 and standard deviation of 3. If it takes 2 minutes for each person to order, and 10 minutes from ordering to getting your food, what is the likelihood that you have at least 15 minutes left to eat your food before you have to go back to class? Assume you have one hour for lunch, and ignore travel time to and from La Panaderia.

    -Answer: .6915

```python
mean = 15 * 2 
std_dev = 3 * 2

total_lunch_time = 60
order_time = 2
eat_time = 15
cook_time = 10

max_wait = total_lunch_time - order_time - eat_time - cook_time
max_wait
```

```python
stats.norm(mean, std_dev).cdf(max_wait)
```

7. Connect to the employees database and find the average salary of current employees, along with the standard deviation. For the following questions, calculate the answer based on modeling the employees salaries with a normal distribution defined by the calculated mean and standard deviation then compare this answer to the actual values present in the salaries dataset.

```python
from pydataset import data
from env import host, username, password
```

```python
url = f'mysql+pymysql://{username}:{password}@{host}/employees'
salaries = pd.read_sql('SELECT * FROM salaries where to_date > now();',url)
```

```python
salaries.head()
```

```python
mean = salaries.salary.mean()
mean
```

```python
std_dev = salaries.salary.std()
```

    - What percent of employees earn less than 60,000?

```python
stats.norm(mean, std_dev).cdf(60000)
```

    - What percent of employees earn more than 95,000?

```python
stats.norm(mean, std_dev).sf(95000)
```

    - What percent of employees earn between 65,000 and 80,000?

```python
more_than = stats.norm(mean, std_dev).sf(65000)
less_than = stats.norm(mean, std_dev).cdf(80000)
more_than, less_than
#high value minus low will give you the range.
```

```python
#another way:
np.diff(stats.norm(mean, std_dev).cdf([65000, 80000]))
```

    - What do the top 5% of employees make?

```python
stats.norm(mean, std_dev).isf(0.05)
```

```python
stats.norm(mean, std_dev).ppf(0.95)
```
